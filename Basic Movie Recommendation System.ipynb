### Weighted Average Method
### W = (Rv+Cm)/(v+m), W=Weighted Rating, R=Avg Rating,v=no. of voted for movie, m=minimum votes required
###                                                                                                                                            for top 250(currently 3000)
###                                    C=Mean vote across the whole report

```python
v = data_cleaned['vote_count']
R = data_cleaned['vote_average']
C = data_cleaned['vote_average'].mean()
m = data_cleaned['vote_count'].quantile(0.70)
```

```python
data_cleaned['weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
data_cleaned
```

```python
data_cleaned.isnull().sum()
```

```python
data_sorted = data_cleaned.sort_values('weighted_average',ascending=False)
```

```python
data_sorted[['original_title','vote_count','vote_average','weighted_average','popularity']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x=data_sorted['weighted_average'].head(10),y=data_sorted['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best Movies by vote',weight='bold')
plt.ylabel('Movies',weight='bold')
plt.xlabel('Ratings',weight='bold')
```

### Recommendation on popularity and weighted average 50%

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
scaled = scale.fit_transform(data_cleaned[['weighted_average','popularity']])
```

```python
data_scaled = pd.DataFrame(scaled,columns=['weighted_average','popularity'])
```

```python
data_scaled.head()
```

```python
data_cleaned[['normalized_weight','normalized_popularity']] = data_scaled
```

```python
data_cleaned.head()
```

```python
data_cleaned['score'] = (data_cleaned['normalized_popularity']*0.5 + data_cleaned['normalized_weight']*0.5)*10
```

```python
data_cleaned.sort_values('score',ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=data_cleaned.sort_values('score',ascending=False)['score'].head(10),
            y=data_cleaned.sort_values('score',ascending=False)['original_title'].head(10))
plt.xlim(0,10)
plt.title('Best Movies by popularity and average rating',weight='bold')
plt.ylabel('Movies',weight='bold')
plt.xlabel('Score',weight='bold')
```
