```python
def collaboritive(name):    
    if(name not in pivot_table.index):
        return 'Sorry No Recommendation'
    query = list(pivot_table.index).index(name)
    distances, indices = model.kneighbors(pivot_table.iloc[query,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('You Might Also Like\n')
        else:
            print('{0}: {1}:'.format(i, pivot_table.index[indices.flatten()[i]], distances.flatten()[i]))
```

### Content Based Filtering

```python
movies.head()
```

```python
genres = []
for i in movies.genres:
    for j in i.split('|'):
        genres.append(j)
genres = list(set(genres))
genres
```

```python
pivot_table1 = pd.DataFrame(index=movies.title, columns=genres).fillna(0)
```

```python
movies.iloc[0,:]
```

```python
for i in movies.values:
    for j in i[2].split('|'):
        if(j in pivot_table1.columns):
            pivot_table1.loc[i[1],[j]] = 1
```

```python
pivot_table1.head()
```

```python
from scipy.sparse import csr_matrix
sparsematrix = csr_matrix(pivot_table1.values)
```

```python
from sklearn.neighbors import NearestNeighbors as knn
model1 = knn(metric='cosine',algorithm='brute')
model1.fit(sparsematrix)
def content():    
    c = 0
    f = 1
    print('enter a movie name or type next to continue or quit to exit','\n')
    for i in pivot_table1.index:
        f = 1
        print(i)
        c += 1
        if(c==10):
            while(True):
                query1 = input()
                if(query1 == 'next'):
                    c = 0
                    f = 0
                    break
                if(query1=='quit'):
                    return
                elif(query1 not in pivot_table1.index):
                    print('\n','Wrong Input')
                    continue
                else:
                    break
            if(f==1):
                break
            else:
                continue

    query = list(pivot_table1.index).index(query1)
    distances, indices = model1.kneighbors(pivot_table1.iloc[query,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Recommendations for {0}:\n'.format(pivot_table1.index[query]))
        else:
            print('{0}: {1}:'.format(i, pivot_table1.index[indices.flatten()[i]]))
```

```python
query = list(pivot_table1.index).index('GoldenEye (1995)')
distances, indices = model1.kneighbors(pivot_table1.iloc[query,:].values.reshape(1,-1),n_neighbors=31)

for i in range(0, len(distances.flatten())):
    string = collaboritive(pivot_table1.index[indices.flatten()[i]])
    if (string == 'Sorry No Recommendation'):
        continue
```
